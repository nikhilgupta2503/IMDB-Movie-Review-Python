# Average Rateing By each Directors

```python
data.columns
```

```python
top_20 = up_data.groupby('Director')['Rating'].mean().sort_values(ascending=False).head(20).reset_index()
bottom_20 = up_data.groupby('Director')['Rating'].mean().sort_values(ascending=True).head(20).reset_index()
```

```python
axt = sns.barplot(x="Director", y="Rating", data=top_20)
plt.xticks(rotation=90)
plt.title("Top 20 Directors by Average Rating")

# Add bar values to each bar
for bar in axt.containers:
    axt.bar_label(bar, fmt='%.2f', fontsize=7, label_type='edge', color='green', weight='light')

plt.show()
```

```python
ax = sns.barplot(x="Director", y="Rating", data=bottom_20)
plt.xticks(rotation=90)
plt.title("Bottom 20 Directors by Average Rating")
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', 
                fontsize=8, color='black', xytext=(0, 5), textcoords='offset points')

plt.show()
```

# Classify Movie On The Based Of Rating

```python
data.columns
```

```python
def rating_category(rating):
    if rating >= 8:
        return "Excellent"
    elif rating >= 6.5:
        return "Good"
    elif rating >= 4.5:
        return "Average"
    else:
        return "Poor"

up_data['Rating_Category'] = up_data['Rating'].apply(rating_category)
```
